## データの可視化について
## ノードグラフ編



ノードグラフは下記2つの要素を組み合わせて構成される。
- ノード：node(vertex) -  点（人、物、場所）
- エッジ：edge  - 辺(繋がり、経路)　
各ノード間の関係性を可視化することができる。

### ノードグラフによるデータの可視化

#### matplotlibを使用した場合

##### 無向グラフ

```python
#描画に必要なライブラリのinstall
!pip install networkx
```

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph() #無向グラフインスタンス生成
H = nx.path_graph(10) #10個のノードpathが生成される
G.add_nodes_from(H) #上記作成したノードパスを追加する
G.add_edges_from([(0, 1), (1, 2)]) #一部ノードにエッジを追加する

nx.draw(G, with_labels=True) #描画する対象及びラベル付加の選択を行う
plt.show(G) 
```

##### 有向グラフ

```python
G = nx.DiGraph() #有向グラフインスタンス生成
G.add_nodes_from(['a','b','c','d']) #a-dのノードを作成
G.add_edges_from([('a','b'),('a','c'),('a','d'),('b','c'),('b','d'),('d','c')])
nx.draw(G, with_labels=True) #描画する対象及びラベル付加の選択を行う
plt.show(G)
```
