## ワードクラウド編



文章中で出現頻度が高い単語を複数選び出し、その頻度に応じた大きさで図示する手法である。  
WordCloudというライブラリを用いることで、英文の文章であればす簡単にワードクラウドを作成することができる。  
※正確には単語の区切りを空白文字(スペース)としてワードクラウドを作成している。

ガスセンサーの時系列データ（https://archive.ics.uci.edu/ml/machine-learning-databases/00309/ ）である  
以下のコードでは、CSVファイルとしてダウンロードしたデータをPandasのDataFrameとして読み込んでいる

```python
!pip install wordcloud
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

#inputとするtexeの定義
text = "Chief Justice Roberts, President Carter, President Clinton, President \
        Bush, President Obama, fellow Americans, and people of the world: \
        thank you. We, the citizens of America, are now joined in a great \
        national effort to rebuild our country and to restore its promise for \
        all of our people. \
        Together, we will determine the course of America and the world for \
        years to come. \
        We will face challenges. We will confront hardships. But we will get \
        the job done. \
        Every four years, we gather on these steps to carry out the orderly \
        and peaceful transfer of power, and we are grateful to President Obama \
        and First Lady Michelle Obama for their gracious aid throughout this \
        transition. They have been magnificent."

#オプションの詳細は下記【参考】を参照のこと
wordcloud = WordCloud(background_color="white",
    width=800,height=600).generate(text)

plt.figure(figsize=(15,12)) #描画サイズの調整
plt.imshow(wordcloud) #描画対象の選択
plt.axis("off") #軸を表示するかどうかの選択
plt.show() #描画
```

【補足】  
ワードクラウドを特定の形で縁取りしたい場合は下記のように設定する

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

#画像ファイルを取り込むために下記2ライブラリを使用する
import numpy as np
from PIL import Image

text = "Chief Justice Roberts, President Carter, President Clinton, President \
        Bush, President Obama, fellow Americans, and people of the world: \
        thank you. We, the citizens of America, are now joined in a great \
        national effort to rebuild our country and to restore its promise for \
        all of our people. \
        Together, we will determine the course of America and the world for \
        years to come. \
        We will face challenges. We will confront hardships. But we will get \
        the job done. \
        Every four years, we gather on these steps to carry out the orderly \
        and peaceful transfer of power, and we are grateful to President Obama \
        and First Lady Michelle Obama for their gracious aid throughout this \
        transition. They have been magnificent."

imagepaht = "./input/star.png" #縁取りしたい形の画像ファイル格納パスを記載する
img_color = np.array(Image.open( imagepaht )) #画像ファイルを配列に変換する
wordcloud = WordCloud(background_color="white",mask = img_color, #maskで縁取りしたい画像ファイルの配列を指定する
                      width=800,height=600).generate(text)

plt.figure(figsize=(15,12)) #描画サイズの調整
plt.imshow(wordcloud) #描画対象の選択
plt.axis("off") #軸を表示するかどうかの選択
plt.show() #描画
```
